# Visualización de los productos más vendidos

Creamos un gráfico para la resolución de los productos más populares para atraer al cliente potencial.

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import wordcloud as w
from collections import Counter
from PIL import Image
from wordcloud import ImageColorGenerator
from wordcloud import WordCloud
```

```python
# Importamos el dataframe madre.
df = pd.read_csv('material.csv')
df.columns
```

```python
# Creamos la columna de la descripción total por producto concatenando cada columna con caracteristicas.

df['Productos.Workcloud'] = df['Especificaciones.calibreECT'].astype(str) + 'ECT'+ ' ' + df['Especificaciones.estructura'] + ' ' + df['Especificaciones.caraExterior']
```

```python
# Convertimos la columna generada a un diccionario que tiene por llaves los valores unicos y como valor la frecuencia de cada uno.

cloud = Counter(df['Productos.Workcloud'])
```

```python
# Creamos la nube usando el metodo WordCloud().generate_from_frequencies()

wordcloud = WordCloud(background_color="black", \
               max_words=20000, max_font_size=10000,\
               random_state=42, width=700, collocations=False, \
               height=400, colormap="YlOrBr").generate_from_frequencies(cloud)

plt.figure(figsize=(25, 25))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
wordcloud.to_file("../recursos/formatosPopulares.png")
```
