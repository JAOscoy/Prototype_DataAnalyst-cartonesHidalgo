# Analisis de producto y desperdicio

Creamos un analisis de los productos vendidos hasta ahora en frecuencia y eficiencia.

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('material.csv')
df.columns
```

```python
from wordcloud import WordCloud
```

```python
df['Productos.Workcloud'] = df['Especificaciones.calibreECT'].astype(str) + 'ECT'+ ' ' + df['Especificaciones.estructura'] + ' ' + df['Especificaciones.caraExterior']
```

```python
grouped_by_title = df.groupby('_id')['Productos.Workcloud'].max()
grouped_by_title
```

```python
tokenized = grouped_by_title.apply(nltk.word_tokenize)
all_words = tokenized.sum()
```

```python
wordcloud = WordCloud(max_font_size=150, background_color="white").generate(' '.join(grouped_by_title))

plt.figure(figsize=(15, 15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
